```python
# CÓDIGO FEITO PARA IMPORTAR DADOS DE PREÇOS HISTÓRICOS DE ATIVOS FINANCEIROS
# (DATA, HORA, ABERTURA, MÁXIMA, MÍNIMA, FECHAMENTO E VOLUME NEGOCIADO) DO
# YAHOO FINANCE DIRETO PARA O METATRADER FIZ ISSO PARA REALIZAR BACKTESTS


from google.colab import drive
drive.mount('/content/drive')
```

```python
# A ORDEM DOS DADOS DEVE SER A SEGUINTE ---> Date, Time, Open, High, Low, Close, Volume

import yfinance as yf
import pandas as pd # Importa o pandas explicitamente

# Definindo o ticker e o período desejado (VER O NOME NO YAHOO)
# O CÓDIGO DO PAPEL PODE SER ENCONTRADO NO https://finance.yahoo.com/
# POR EXEMPLO VALE3 DEVE-SE ESCREVER VALE3.SA
ticker = input("Digite o código do papel (max period): ")
periodo = "max"

# Baixando os dados históricos e armazenando em um DataFrame
df = yf.download(ticker, period=periodo)
```

```python
df.head()
```

```python
# COMO

# Assuming the first column is the index and contains dates
# Reset the index to make 'Date' a column
df = df.reset_index()

# Rename the 'Date' column for clarity if its not named 'Date'
# df = df.rename(columns={df.columns[0]: 'Date'})
# Remove the values beyond the 10th character in the 'Date' column
df['Date'] = df['Date'].astype(str).str[:10]

# If you need 'Date' as index again:
# df = df.set_index('Date')
```

```python
df
```

```python
# prompt: Using dataframe df: Preciso que adiciona uma coluna entre a primeira coluna e a segunda coluna, que tenha a seguinte informação (00:00:00)

# Insert a new column named 'Time' at the second position (index 1)
# The new column will contain the value '00:00:00' for all rows

df.insert(0, 'Time', '00:00:00')
df.head()
```

```python
# CASO VENHA A COLUNA ADJ CLOSE #
# prompt: Using dataframe df: Preciso excluir a terceira coluna de valores

# Assuming 'df' is your DataFrame
# Delete the third column, which is 'Adj Close' based on the provided information

# Check if the column exists before deleting
if 'Adj Close' in df.columns:
    df = df.drop(columns=['Adj Close'])
    print("Column 'Adj Close' deleted successfully.")
else:
    print("Column 'Adj Close' not found in the DataFrame.")

df.head()
```

```python
# Colocar na ordem OHLC
# prompt: As colunas do df são Date, Time, Close, High, Low, Open, Volume. Preciso que fiquem na seguinte ordem: Date, Time, Open, High, Low, Close, Volume

# Reset the index to make 'Date' a regular column
df = df.reset_index()

# Reorder the columns
df = df[['Date', 'Time', 'Open', 'High', 'Low', 'Close', 'Volume']]

df.head()
```

```python
# prompt: Using dataframe df: Preciso que troca de lugar a sexta coluna com a terceira coluna

# Get the name of the columns
cols = df.columns.tolist()

# Swap the positions of the 6th (index 5) and 3rd (index 2) columns
cols[1], cols[4] = cols[4], cols[1]

# Reorder the DataFrame based on new column order
df = df[cols]

# Print the updated DataFrame
df
```
